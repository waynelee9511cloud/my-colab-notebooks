## ✅ 步驟 3: 驗證專案結構

確認所有必要的檔案都在正確的位置。

```python
import os
from pathlib import Path

# 檢查必要的檔案
required_files = [
    'web_interface.py',
    'modules/protocol_parser.py',
    'modules/crf_generator.py',
    'modules/dvp_generator.py',
    'modules/user_guide_generator.py',
    'modules/word_formatter.py',
]

print("檢查專案檔案...\n")
all_exists = True

for file in required_files:
    exists = Path(file).exists()
    status = "✓" if exists else "✗"
    print(f"{status} {file}")
    if not exists:
        all_exists = False

print()
if all_exists:
    print("✅ 所有必要檔案都存在！")
else:
    print("❌ 部分檔案缺失，請確認專案結構")
```

## 📝 使用說明

啟動 Web UI 後，請按照以下步驟操作：

### 1. 設定 API Key
- 訪問 [Google AI Studio](https://makersuite.google.com/app/apikey)
- 登入並創建 API Key
- 複製 API Key 並貼到 Web UI 的輸入框中
- 點擊「設定 API Key」按鈕

### 2. 上傳檔案
- **Protocol PDF (必填)**: 上傳您的臨床試驗 Protocol PDF 檔案
- **公司 Logo (選填)**: 上傳公司 Logo 圖片（PNG、JPG 格式）

### 3. 解析 Protocol
- 點擊「開始解析 Protocol」按鈕
- 等待 AI 分析（約 30-60 秒）
- 系統會自動提取 Protocol 中的關鍵資訊

### 4. 編輯資訊（可選）
- 在 JSON 編輯器中查看提取的資訊
- 如需修改，直接編輯 JSON 內容
- 點擊「更新資訊」按鈕保存修改

### 5. 選擇文件類型
- 勾選您想要生成的文件類型：
  - **CRF**: 病例報告表
  - **DVP**: 資料驗證計劃
  - **User Guide**: 使用者指南
  - **DMP**: 資料管理計劃

### 6. 生成文件
- 點擊「一鍵生成文件」按鈕
- 等待生成完成
- 在「下載生成的文件」區域點擊下載按鈕

## 🐛 故障排除

### 問題 1: 模組導入錯誤
**解決方法**: 確認所有必要的檔案都已上傳，並執行步驟 3 的驗證。

### 問題 2: API Key 無效
**解決方法**: 
- 確認 API Key 是否正確
- 檢查 API Key 是否已啟用
- 訪問 [Google AI Studio](https://makersuite.google.com/app/apikey) 重新生成

### 問題 3: PDF 解析失敗
**解決方法**:
- 確認 PDF 檔案未加密
- 檢查 PDF 檔案大小（建議小於 50 MB）
- 確認 PDF 內容可以被複製（有些掃描版 PDF 無法提取文字）

### 問題 4: 文件生成失敗
**解決方法**:
- 確認 Protocol 資訊已正確提取
- 檢查輸出目錄權限
- 查看錯誤訊息並根據提示操作

### 問題 5: Web UI 無法訪問
**解決方法**:
- 在 Colab 中重新執行啟動代碼
- 確認防火牆沒有阻擋連線
- 嘗試使用不同的瀏覽器

## 💡 提示和技巧

1. **提高解析準確度**:
   - 使用高品質的 PDF 檔案
   - 確保 Protocol 結構清晰
   - 解析後仔細檢查提取的資訊

2. **優化生成速度**:
   - 一次只生成必要的文件類型
   - 使用 `gemini-1.5-flash` 模型（更快）
   - 確保網路連線穩定

3. **自訂文件內容**:
   - 在 JSON 編輯器中修改提取的資訊
   - 添加額外的納入/排除標準
   - 調整訪視時程

4. **批次處理**:
   - 對於多個 Protocol，可以重複使用同一個 Web UI 實例
   - 每次處理完後，重新上傳新的 PDF 即可
